##     19. What is the difference between the following two lines? Which one will give a larger value? Will this be the case for other texts?

    >>> sorted(set([w.lower() for w in text1]))
    >>> sorted([w.lower() for w in set(text1)])

### first line will give smaller value in general.
Let explain the second line first
* the second line will first get all unique word(case-sensitive), which means words like "do" and "Do" are different words, which will be in the set. Then it lowercases each words, so there are two "do", one from "do", and one from "Do".
* the first line goes the opposite way. It lowercases each and every word so that 'Do' becomes 'do', then they are turned into a set, which means a list of all unique words.

##     28. Define a function percent(word, text) that calculates how often a given word occurs in a text and expresses the result as a percentage.
