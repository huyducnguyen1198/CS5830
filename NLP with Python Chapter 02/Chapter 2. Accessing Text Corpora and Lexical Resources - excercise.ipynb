##     Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.


```python
def supergloss(word):
    assert type(word) == nltk.corpus.reader.wordnet.Synset
    
    hyper = word.hypernyms()
    hypo = word.hyponyms()
    wordDef = word.definition()
    hyperDef = [w.definition() for w in hyper]
    hypoDef = [w.definition() for w in hypo]
    
    defi = [wordDef] + hyperDef + hypoDef
    return "".join(defi)


supergloss(word = wn.synsets('dog')[1])
```

##     Write a program to find all words that occur at least three times in the Brown Corpus.


```python
def countOccurence(words):
    d = {}
    for w in words:
        w = w.lower()
        if w in d:
            d[w] += 1
        else:
            d[w] = 1
    return d
occur = countOccurence(brown.words())
occur3 = {k:v for k, v in occur.items() if v >= 3 }
occur3
```

##     Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.



```python
def topWithoutStopwords(stopwords, words, num=50):
    wordsNoStopwords = [w.lower() for w in words if w.lower() not in stopwords]
    freq = nltk.FreqDist(wordsNoStopwords)
    top50 = sorted(freq, key=lambda x: freq[x], reverse=True)[:num]
    return {k: freq[k] for k in top50}
    
stopwords = nltk.corpus.stopwords.words("english")
words = brown.words()
topWithoutStopwords(stopwords, words, 50)
```

##     Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.



```python
def topFreqBigramWithoutStopword(stopwords, words, num=50):
    words = [w.lower() for w in words]
    bi = nltk.bigrams(words)
    biNoStop = []
    for b in bi:
        if b[0] in stopwords or b[1] in stopwords:
            continue
        else:
            biNoStop.append(b)
    freq = nltk.FreqDist(biNoStop)
    top50 = sorted(freq, key=lambda x: freq[x], reverse=True)[:num]
    return {k: freq[k] for k in top50}
topFreqBigramWithoutStopword(stopwords, words)
```

##     Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary.


```python
cmu = nltk.corpus.cmudict
def countPron(words):
    d = cmu.dict()
    pronCount = 0
    for w in words:
        if w in d:
            pronCount += len(d[w][0])
    return pronCount
print("num of word", len(brown.words()))
pronCount = countPron(brown.words())
print("num of pronunciation: ", pronCount )
print("average pron: ", pronCount/len(brown.words()))
```

##     Define a function hedge(text) that processes a text and produces a new version with the word 'like' between every third word.


```python
def hedge(text):
    token = text.split()
    newVer = []
    for i in range(len(token)):
        if i%3 == 2:
            newVer.append("like")
        newVer.append(token[i])
    return " ".join(newVer)
text = brown.raw("ca01")
hedge(text)
```


##     ● The polysemy of a word is the number of senses it has. Using WordNet, we can determine that the noun dog has seven senses with len(wn.synsets('dog', 'n')). Compute the average polysemy of nouns, verbs, adjectives, and adverbs according to WordNet.


```python

for wType in [wn.VERB, wn.NOUN, wn.ADJ, wn.ADV]:
    nouns = wn.all_synsets(wType)
    n = 0
    count = 0
    for i in nouns:
        n += 1
        name = i.name().split(".")[0]
        #get all senses
        senses = wn.synsets(name, wType)
        count += len(senses)
    print(wType, ": ",count / n)
    
```

##  ● What is the branching factor of the noun hypernym hierarchy? I.e., for every noun synset that has hyponyms—or children in the hypernym hierarchy—how many do they have on average? You can get all noun synsets using wn.all_synsets('n').


```python
nouns = wn.all_synsets('n')
n = 0
count = 0
for i in nouns:
    n += 1
    numHypo = len(i.hyponyms())
    if numHypo > 0:
        count += numHypo
    
print('Branching factor: ', count/n)
```

## ● Define a function find_language() that takes a string as its argument and returns a list of languages that have that string as a word. Use the udhr corpus and limit your searches to files in the Latin-1 encoding.


```python
def find_language(string):
    udhr = nltk.corpus.udhr
    langList = [l for l in udhr.fileids() if 'Latin1' in l]
    searchLang = []
    for l in langList:
        lang = udhr.words(l)
        lang = [l.lower() for l in lang]
        if string in lang:
            searchLang.append(l)
    return searchLang
find_language("to")
```
