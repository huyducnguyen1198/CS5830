## ○ We can use the slice notation to remove morphological endings on words. For example, 'dogs'[:-1] removes the last character of dogs, leaving dog. Use slice notation to remove the affixes from these words (we’ve inserted a hyphen to indicate the affix boundary, but omit this from your strings): dish-es, run-ning, nation-ality, un-do, pre-heat.

```python
affix = ['es', 'ning', 'ality', 'un', 'pre']
words = ['dishes', 'running', 'nationality', 'undo', 'preheat']
newWords = []
for w in words:
    index, *_ = [(w.index(a), len(a)) for a in affix if a in w]
    if index[0] == 0:
        newWords.append(w[index[1]:])
    else:
        newWords.append(w[:index[0]])
        
newWords
```

## Write a function unknown() that takes a URL as its argument, and returns a list of unknown words that occur on that web page. In order to do this, extract all substrings consisting of lowercase letters (using re.findall()) and remove any items from this set that occur in the Words Corpus (nltk.corpus.words). Try to categorize these words manually and discuss your findings.


```python
import regex as re
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun if unknown

# Create an instance of the LancasterStemmer
lemmatizer = WordNetLemmatizer()
porter = PorterStemmer()
html = urlopen('https://en.wikipedia.org/wiki/Spacetime').read()
soup = BeautifulSoup(html, 'html.parser')
raw = soup.get_text()
words = nltk.word_tokenize(raw)

reWords = re.findall(r'\b[a-z]+\b', raw)

wordCorpus = nltk.corpus.words
wordCorpus = wordCorpus.words('en')

# Tag words with POS
tagged_words = nltk.pos_tag(reWords)

# Lemmatize words with correct POS
lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in tagged_words]


#reWords = [lemmatizer.lemmatize(w) for w in reWords]
#reWords = [porter.stem(w) for w in reWords]

reWords = set(lemmatized_words)

unknown = []

for w in reWords:
    if w not in wordCorpus:
        unknown.append(w)
print((unknown))
```
